```python
file_path = '/Users/aagneshshifak/Downloads/data.csv'
import pandas as pd
data = pd.read_csv(file_path)
import matplotlib.pyplot as plt

goals_data = data.dropna(subset=["Type"])
goals_per_season = goals_data["Season"].value_counts().sort_index()

plt.figure(figsize=(12, 6))
goals_per_season.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Lionel Messi: Goals Per Season", fontsize=16)
plt.xlabel("Season", fontsize=12)
plt.ylabel("Number of Goals", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/Users/aagneshshifak/Downloads/data.csv')

goal_types = [
    'Left-footed shot', 'Right-footed shot', 'Header', 'Solo run', 'Penalty',
    'Deflected shot on goal', 'Direct free kick', 'Penalty rebound',
    'Counter attack goal', 'Chest', 'Tap-in', 'Long distance kick'
]
goals_data = data[data['Type'].isin(goal_types)].copy()

def convert_minute(minute):
    if pd.isnull(minute):
        return None
    if isinstance(minute, str):
        if '+' in minute:
            base, extra = minute.split('+')
            return int(base) + int(extra)
        elif minute.isdigit():
            return int(minute)
    return int(minute) if isinstance(minute, (int, float)) else None

goals_data.loc[:, 'Minute_numeric'] = goals_data['Minute'].apply(convert_minute)
goals_data = goals_data.dropna(subset=['Minute_numeric'])

plt.figure(figsize=(10, 6))
plt.hist(goals_data['Minute_numeric'], bins=30, color='skyblue', edgecolor='black', alpha=0.8)
plt.title("Histogram of Goal Scoring Minutes", fontsize=16)
plt.xlabel("Minute", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
